```python
covariance_matrix = iris[['sepal_length', 'sepal_width']].corr()

print(covariance_matrix)
```

```python

plt.figure(figsize=(10, 6))
scatter = plt.scatter(iris['sepal_length'], iris['sepal_width'], alpha=0.5)

# Adding the correlation coefficient to the plot
corr_coef = iris['sepal_length'].corr(iris['sepal_width'])
plt.title(f'Scatter plot of Sepal Length vs Sepal Width\nCorrelation Coefficient: {corr_coef:.2f}')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')

# Show the plot
plt.show()
```
